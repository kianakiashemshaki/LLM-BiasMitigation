```python
import re

def preprocess_dataset(df):
    df.columns = [col.strip().lower() for col in df.columns]

    df = df.dropna(subset=['gender', 'hard_text', 'profession'])

    df['hard_text'] = (
        df['hard_text']
        .str.lower()
        .str.strip()
        .apply(lambda x: re.sub(r'\s+', ' ', x))
        .apply(lambda x: re.sub(r'[^a-z\s]', '', x))
    )

    df = df.reset_index(drop=True)

    return df

df_train = preprocess_dataset(df_train)
df_dev = preprocess_dataset(df_dev)
df_test = preprocess_dataset(df_test)

print(f"Train shape: {df_train.shape}")
print(f"Dev shape: {df_dev.shape}")
print(f"Test shape: {df_test.shape}")

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot(dfs, column, title):
    fig, axes = plt.subplots(1, len(dfs), figsize=(15, 5), sharey=True)
    dataset_names = ['Train', 'Dev', 'Test']

    for ax, df, name in zip(axes, dfs, dataset_names):
        sns.countplot(data=df, x=column, order=df[column].value_counts().index, palette="viridis", ax=ax)
        ax.set_title(f"{name}")
        ax.set_xlabel('')
        ax.set_ylabel('Count')
        ax.tick_params(axis='x', rotation=45)

    fig.suptitle(title, fontsize=16)
    plt.tight_layout()
    plt.show()

plot([df_train, df_dev, df_test], 'gender_text', 'Gender Distribution by Dataset')


def plot_operation(dfs, column, title):
    fig, axes = plt.subplots(1, len(dfs), figsize=(18, 12), sharey=True)
    dataset_names = ['Train', 'Dev', 'Test']

    for ax, df, name in zip(axes, dfs, dataset_names):
        order = df[column].value_counts().index
        sns.countplot(data=df, y=column, order=order, palette="viridis", ax=ax)
        ax.set_title(f"{name}")
        ax.set_xlabel('Count')
        ax.set_ylabel('')

    fig.suptitle(title, fontsize=18)
    plt.tight_layout()
    plt.show()

plot_operation(
    [df_train, df_dev, df_test],
    'profession_text',
    'Profession Distribution (Horizontal)'
)

```

```python
def class_distribution_report(dfs, dataset_names, columns):
    for col in columns:
        print(f"\n{'='*20} {col} Distribution {'='*20}")
        for df, name in zip(dfs, dataset_names):
            counts = df[col].value_counts()
            percentages = df[col].value_counts(normalize=True) * 100
            report_df = pd.DataFrame({'Count': counts, 'Percentage': percentages.round(2)})
            print(f"\n{name}:\n{report_df}")

dfs = [df_train, df_dev, df_test]
dataset_names = ['Train', 'Dev', 'Test']
columns = ['gender_text', 'profession_text']

class_distribution_report(dfs, dataset_names, columns)
```

```python
from imblearn.over_sampling import RandomOverSampler
import pandas as pd

def oversample_df(df, target_col, random_state=42):

    ros = RandomOverSampler(random_state=random_state)
    X_res, y_res = ros.fit_resample(df, df[target_col])
    return pd.DataFrame(X_res, columns=df.columns)

df_train_gender_balanced = oversample_df(df_train, target_col='gender_text')
df_train_profession_balanced = oversample_df(df_train, target_col='profession_text')

print("=== Gender balance ===")
print("before:\n", df_train['gender_text'].value_counts())
print("after:\n", df_train_gender_balanced['gender_text'].value_counts())

print("\n=== Profession balance ===")
print("before:\n", df_train['profession_text'].value_counts())
print("after:\n", df_train_profession_balanced['profession_text'].value_counts())
```

```python
import torch
from transformers import BertTokenizer
from tqdm.auto import tqdm

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def build_label_map(labels):
    unique_labels = sorted(set(labels))
    return {label: idx for idx, label in enumerate(unique_labels)}

def tokenize_dataset(df, text_column, label_column, tokenizer, max_length=128, batch_size=32, main_pbar=None, desc="Tokenizing"):

    label_map = build_label_map(df[label_column])
    labels = torch.tensor([label_map[lbl] for lbl in df[label_column]], dtype=torch.long)

    all_input_ids = []
    all_attention_masks = []

    for i in tqdm(range(0, len(df), batch_size), desc=desc, leave=True):
        batch_texts = df[text_column].iloc[i:i+batch_size].tolist()
        encodings = tokenizer(
            batch_texts,
            truncation=True,
            padding='max_length',
            max_length=max_length,
            return_tensors='pt'
        )
        all_input_ids.append(encodings['input_ids'])
        all_attention_masks.append(encodings['attention_mask'])

    if main_pbar is not None:
        main_pbar.update(1)

    input_ids = torch.cat(all_input_ids, dim=0)
    attention_masks = torch.cat(all_attention_masks, dim=0)

    return {
        'input_ids': input_ids,
        'attention_mask': attention_masks,
        'labels': labels
    }

steps = [
    ("Gender Train", df_train_gender_balanced, 'hard_text', 'gender_text'),
    ("Gender Dev", df_dev, 'hard_text', 'gender_text'),
    ("Gender Test", df_test, 'hard_text', 'gender_text'),
    ("Profession Train", df_train_profession_balanced, 'hard_text', 'profession_text'),
    ("Profession Dev", df_dev, 'hard_text', 'profession_text'),
    ("Profession Test", df_test, 'hard_text', 'profession_text')
]

encodings = {}
with tqdm(total=len(steps), desc="Overall Tokenization", leave=True) as main_pbar:
    for name, df, text_col, label_col in steps:
        encodings[name] = tokenize_dataset(
            df,
            text_column=text_col,
            label_column=label_col,
            tokenizer=tokenizer,
            max_length=128,
            batch_size=32,
            main_pbar=main_pbar,
            desc=f"{name} ({label_col})"
        )

print("\nShapes for Gender classification:")
for split in ["Gender Train", "Gender Dev", "Gender Test"]:
    print(split, encodings[split]['input_ids'].shape)

print("\nShapes for Profession classification:")
for split in ["Profession Train", "Profession Dev", "Profession Test"]:
    print(split, encodings[split]['input_ids'].shape)
```

```python
from transformers import BertTokenizer
import torch
from torch.utils.data import TensorDataset
from tqdm.auto import tqdm

CONFIG = {
    "MAX_SAMPLES": 2000,
    "BATCH_SIZE": 16,
    "MAX_LENGTH": 64
}

tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

def build_label_map(labels):
    unique_labels = sorted(set(labels))
    return {label: idx for idx, label in enumerate(unique_labels)}

def tokenize_dataset(df, text_column, label_column, max_samples=CONFIG["MAX_SAMPLES"]):
    df = df.iloc[:max_samples]
    label_map = build_label_map(df[label_column])
    labels = torch.tensor([label_map[lbl] for lbl in df[label_column]], dtype=torch.long)

    input_ids_list = []
    attention_mask_list = []

    for text in tqdm(df[text_column], desc=f"Tokenizing {label_column}"):
        enc = tokenizer(text, truncation=True, padding='max_length', max_length=CONFIG["MAX_LENGTH"], return_tensors='pt')
        input_ids_list.append(enc['input_ids'])
        attention_mask_list.append(enc['attention_mask'])

    input_ids = torch.cat(input_ids_list, dim=0)
    attention_mask = torch.cat(attention_mask_list, dim=0)

    return {
        'input_ids': input_ids,
        'attention_mask': attention_mask,
        'labels': labels
    }

encodings = {}
encodings["Gender Train"] = tokenize_dataset(df_train_gender_balanced, "hard_text", "gender_text")
encodings["Gender Dev"] = tokenize_dataset(df_dev, "hard_text", "gender_text")
encodings["Gender Test"] = tokenize_dataset(df_test, "hard_text", "gender_text")

encodings["Profession Train"] = tokenize_dataset(df_train_profession_balanced, "hard_text", "profession_text")
encodings["Profession Dev"] = tokenize_dataset(df_dev, "hard_text", "profession_text")
encodings["Profession Test"] = tokenize_dataset(df_test, "hard_text", "profession_text")

print("Tokenization done. Shapes:")
for k, v in encodings.items():
    print(k, v['input_ids'].shape)
```

```python
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset, RandomSampler, SequentialSampler
from transformers import BertForSequenceClassification
from sklearn.metrics import accuracy_score, f1_score
from fairlearn.metrics import MetricFrame, demographic_parity_difference, equalized_odds_difference
import pandas as pd
from tqdm.auto import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

CONFIG = {
    "EPOCHS": 2,
    "BATCH_SIZE": 16,
    "LR": 2e-5,
    "MAX_SAMPLES": 2000
}

def build_loader(encodings, batch_size=CONFIG["BATCH_SIZE"], shuffle=False):
    dataset = TensorDataset(encodings['input_ids'], encodings['attention_mask'], encodings['labels'])
    sampler = RandomSampler(dataset) if shuffle else SequentialSampler(dataset)
    return DataLoader(dataset, sampler=sampler, batch_size=batch_size)

def evaluate_model(model, loader):
    model.eval()
    preds, trues = [], []
    with torch.no_grad():
        for input_ids, attention_mask, labels in tqdm(loader, desc="Evaluating", leave=False):
            input_ids, attention_mask, labels = [x.to(device) for x in (input_ids, attention_mask, labels)]
            outputs = model(input_ids, attention_mask=attention_mask)
            logits = outputs.logits
            preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
            trues.extend(labels.cpu().numpy())
    return preds, trues

def fairness_metrics(y_true, y_pred, sensitive):
    mf = MetricFrame(
        metrics={"accuracy": accuracy_score, "f1": lambda y,t: f1_score(y,t,average="macro")},
        y_true=y_true,
        y_pred=y_pred,
        sensitive_features=sensitive
    )
    return {
        "dpd": demographic_parity_difference(y_true, y_pred, sensitive_features=sensitive),
        "eod": equalized_odds_difference(y_true, y_pred, sensitive_features=sensitive)
    }

def train_model(model, train_loader, dev_loader, sensitive_dev):
    optimizer = torch.optim.Adam(model.parameters(), lr=CONFIG["LR"])
    criterion = nn.CrossEntropyLoss()
    model.to(device)
    for epoch in range(CONFIG["EPOCHS"]):
        model.train()
        total_loss = 0
        for input_ids, attention_mask, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{CONFIG['EPOCHS']}", leave=False):
            input_ids, attention_mask, labels = [x.to(device) for x in (input_ids, attention_mask, labels)]
            optimizer.zero_grad()
            outputs = model(input_ids, attention_mask=attention_mask)
            loss = criterion(outputs.logits, labels)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()
        print(f"Epoch {epoch+1} average loss: {total_loss/len(train_loader):.4f}")
    return model

def run_task(task_name, enc_train, enc_dev, enc_test, sensitive_dev, sensitive_test):

    train_loader = build_loader({k:v[:CONFIG["MAX_SAMPLES"]] for k,v in enc_train.items()}, shuffle=True)
    dev_loader = build_loader({k:v[:CONFIG["MAX_SAMPLES"]] for k,v in enc_dev.items()})
    test_loader = build_loader({k:v[:CONFIG["MAX_SAMPLES"]] for k,v in enc_test.items()})

    num_labels = len(torch.unique(enc_train['labels']))
    model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=num_labels)

    model = train_model(model, train_loader, dev_loader, sensitive_dev)

    preds_test, trues_test = evaluate_model(model, test_loader)
    fair_metrics = fairness_metrics(trues_test, preds_test, sensitive_test)

    return {
        "task": task_name,
        "accuracy": accuracy_score(trues_test, preds_test),
        "macro_f1": f1_score(trues_test, preds_test, average="macro"),
        "dpd": fair_metrics["dpd"],
        "eod": fair_metrics["eod"]
    }

results_list = []

results_list.append(run_task(
    "Gender_Baseline",
    encodings["Gender Train"], encodings["Gender Dev"], encodings["Gender Test"],
    df_dev['gender_text'].iloc[:CONFIG["MAX_SAMPLES"]].tolist(),
    df_test['gender_text'].iloc[:CONFIG["MAX_SAMPLES"]].tolist()
))

df_results = pd.DataFrame(results_list)
print(df_results)
```

```python
from transformers import BertTokenizer
from tqdm.auto import tqdm
import torch

tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

CONFIG["MAX_LENGTH"] = 2000

def build_label_map(labels):
    unique_labels = sorted(set(labels))
    return {label: idx for idx, label in enumerate(unique_labels)}

def tokenize_dataset(df, text_column, label_column, max_samples=None):
    if max_samples:
        df = df.iloc[:max_samples]

    label_map = build_label_map(df[label_column])
    labels = torch.tensor([label_map[lbl] for lbl in df[label_column]], dtype=torch.long)

    input_ids_list = []
    attention_mask_list = []

    for text in tqdm(df[text_column], desc=f"Tokenizing {label_column}"):
        enc = tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=CONFIG["MAX_LENGTH"],
            return_tensors="pt"
        )
        input_ids_list.append(enc['input_ids'])
        attention_mask_list.append(enc['attention_mask'])

    input_ids = torch.cat(input_ids_list, dim=0)
    attention_mask = torch.cat(attention_mask_list, dim=0)

    return {
        "input_ids": input_ids,
        "attention_mask": attention_mask,
        "labels": labels
    }
```

```python
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler
from transformers import BertForSequenceClassification
import torch
from sklearn.metrics import accuracy_score, f1_score
from fairlearn.metrics import MetricFrame, demographic_parity_difference, equalized_odds_difference
from tqdm.auto import tqdm

BATCH_SIZE = 16
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
MAX_SAMPLES = 2000

def build_loader(encodings, batch_size=BATCH_SIZE, shuffle=False):
    dataset = TensorDataset(
        encodings['input_ids'][:MAX_SAMPLES],
        encodings['attention_mask'][:MAX_SAMPLES],
        encodings['labels'][:MAX_SAMPLES]
    )
    sampler = RandomSampler(dataset) if shuffle else SequentialSampler(dataset)
    return DataLoader(dataset, sampler=sampler, batch_size=batch_size)

train_loader_prof = build_loader(encodings["Profession Train"], shuffle=True)
dev_loader_prof   = build_loader(encodings["Profession Dev"])
test_loader_prof  = build_loader(encodings["Profession Test"])

num_prof_labels = len(set(df_train_profession_balanced['profession_text']))
model_prof = BertForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    num_labels=num_prof_labels
)
model_prof.to(DEVICE)

def train_one_epoch(model, loader, optimizer, criterion):
    model.train()
    total_loss = 0
    for batch in tqdm(loader, desc="Training Profession", leave=False):
        input_ids, attention_mask, labels = [x.to(DEVICE) for x in batch]
        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=attention_mask)
        loss = criterion(outputs.logits, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    return total_loss / len(loader)

def evaluate(model, loader):
    model.eval()
    preds, trues = [], []
    for batch in tqdm(loader, desc="Evaluating Profession", leave=False):
        input_ids, attention_mask, labels = [x.to(DEVICE) for x in batch]
        with torch.no_grad():
            outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits
        preds.extend(torch.argmax(logits, dim=1).cpu().numpy())
        trues.extend(labels.cpu().numpy())
    return preds, trues

optimizer_prof = torch.optim.AdamW(model_prof.parameters(), lr=2e-5)
criterion_prof = torch.nn.CrossEntropyLoss()

print("Training Profession model...")
train_loss = train_one_epoch(model_prof, train_loader_prof, optimizer_prof, criterion_prof)
print(f"Avg training loss: {train_loss:.4f}")

preds_test, trues_test = evaluate(model_prof, test_loader_prof)

acc = accuracy_score(trues_test, preds_test)
macro_f1 = f1_score(trues_test, preds_test, average="macro")
print(f"Test Accuracy: {acc:.4f}, Macro F1: {macro_f1:.4f}")

sensitive_test = df_test['profession_text'].iloc[:MAX_SAMPLES].tolist()
mf = MetricFrame(
    metrics={"accuracy": accuracy_score, "f1": lambda y, t: f1_score(y, t, average="macro")},
    y_true=trues_test,
    y_pred=preds_test,
    sensitive_features=sensitive_test
)
dp_diff = demographic_parity_difference(trues_test, preds_test, sensitive_features=sensitive_test)

eod_per_class = {}
num_classes = len(set(trues_test))
for c in range(num_classes):
    y_true_bin = [1 if y==c else 0 for y in trues_test]
    y_pred_bin = [1 if y==c else 0 for y in preds_test]
    eod_per_class[c] = equalized_odds_difference(y_true_bin, y_pred_bin, sensitive_features=sensitive_test)

print("Fairness metrics by group:\n", mf.by_group)
print("Demographic Parity Difference (DPD):", dp_diff)
print("Equalized Odds Difference (EOD) per class:", eod_per_class)
```
